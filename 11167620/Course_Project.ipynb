<h1>Feature Distribution</h1>
The representation will give us how the dataset specifically the <b>alcohol</b> is distributed. From the dataset, we realize that the alcohol content has numerous unique values, and some values keep repeating. We can find that the values between <b><i>9.00<i/></b> and <b><i>10.00<i/></b> have the most occurrences in the dataset especially 9.50.

```python
df['alcohol'].unique()
```

```python
plt.figure(figsize=(12, 7))
plt.hist(df['alcohol'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Alcohol Content in Red Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```
